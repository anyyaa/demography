```python
import matplotlib.pyplot as plt

years = [i for i in range(1939, 1946)]
birth_rates = [38.4, 33.7, 33.3, 20.2, 9.9, 10.5, 10.8]

plt.figure(figsize=(10, 6))
plt.plot(years, birth_rates, marker='o', linestyle='-', color='#34a8eb', linewidth=2)

plt.title('      ( 1000 )', fontsize=14)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(years)
plt.tight_layout()
plt.show()
```

```python

categories = ['  ( 1000)', '  ( 1000)']

birth_rate_1800 = 50
death_rate_1800 = 37.5
birth_rate_2025 = 8.5  
death_rate_2025 = 13.9  

color_1800 = '#34ebba'
color_2025 = '#34a8eb'

values_1800 = [birth_rate_1800, death_rate_1800]
values_2025 = [birth_rate_2025, death_rate_2025]

x = range(len(categories))

plt.figure(figsize=(10, 6))
bar_width = 0.35

bars1 = plt.bar([i - bar_width/2 for i in x], values_1800, width=bar_width, label='~1800 ', color='#34ebba')
bars2 = plt.bar([i + bar_width/2 for i in x], values_2025, width=bar_width, label=' 2025', color='#34a8eb')

for bars in (bars1, bars2):
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval:.1f}', ha='center', va='bottom', fontsize=11)

plt.title('   : ~1800 vs  2025', fontsize=14)
plt.ylabel(' 1000 ', fontsize=12)
plt.xticks(ticks=x, labels=categories, fontsize=11)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
years = [i for i in range(2000, 2024)]
birth_rate = [8.7, 9.0, 9.7, 10.2, 10.4, 10.2, 10.3, 11.3, 12.0, 12.3, 12.5, 12.6, 13.3, 13.2, 13.3, 13.3, 12.9, 11.5, 10.9, 10.1, 9.8, 9.6, 8.9, 8.4]
death_rate = [15.3, 15.6, 16.2, 16.4, 15.9, 16.1, 15.1, 14.6, 14.5, 14.1, 14.2, 13.5, 13.3, 13.0, 13.1, 13.0, 12.9, 12.4, 12.5, 12.3, 14.6, 16.7, 12.9, 12.5]
natural_increase = [b - d for b, d in zip(birth_rate, death_rate)]
print(len(years))
print(len(birth_rate))
# 
x = np.arange(len(years))
width = 0.25

#  
plt.figure(figsize=(12, 6))
plt.bar(x - width, birth_rate, width, label='', color='#34ebba')
plt.bar(x, death_rate, width, label='', color='#34a8eb')
plt.bar(x + width, natural_increase, width, label='/', color=['#34ebba' if v >= 0 else '#eb4034' for v in natural_increase])

# 
plt.xticks(x, years)
plt.ylabel(' 1000 ')
plt.title('    (20002024)')
plt.axhline(0, color='black', linewidth=0.7)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

years = list(range(2013, 2025))
living_wage = [
    6600, 7300, 9662, 9772, 9900, 10038,
    10753, 10843, 11653, 12654, 14375, 17570
]
inflation_rates = [2.71, 9.51, 7.42, 11.92, 8.39, 4.91, 3.05, 4.27, 2.52, 5.38, 12.91, 11.36, 6.45]
x = 1 * (1 + inflation_rates[0] / 100)

inflation_adjusted_wage = [living_wage[0] * (1 + inflation_rates[0] / 100)]
for i in range(1, len(inflation_rates)-1):
    inflation_adjusted_wage.append(inflation_adjusted_wage[-1] * (1 + inflation_rates[i] / 100))
    x *= (1 + inflation_rates[i] / 100)
print(len(inflation_adjusted_wage))
print(x)
plt.figure(figsize=(10, 6))

plt.plot(years, living_wage, marker='o', color='#34ebba', label=' ')

plt.plot(years, inflation_adjusted_wage, marker='o', color='#ff6347', label='    ')

plt.title('    (20132024)\n  ', fontsize=14)
plt.xlabel('')
plt.ylabel('  ')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(years)
plt.ylim(0, max(living_wage) + 5000)
plt.legend()
plt.tight_layout()

plt.show()
```
